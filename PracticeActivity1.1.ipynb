```python
# 1. Bullet points of Group Members
items = ["Joe", "Dom"]

for item in items:
    print(f"• {item}")

# 2. Something we enjoyed for fun (images)
image_urls = [
    "https://assets-prd.ignimgs.com/2022/04/19/wow-dragonflight-button-1-1650398895381.jpg",
    "https://dxbhsrqyrr690.cloudfront.net/sidearm.nextgen.sites/gopoly.com/images/2023/3/17/Demas-NCAA1.jpg",
]

# Captions for the images
captions = [
    "Joe",
    "Dom"
]

# Set number of images and adjust size and fitting into a plot
num_images = len(image_urls)

fig, axes = plt.subplots(1, num_images, figsize=(15, 5))

for ax, img_url, caption in zip(axes, image_urls, captions):
    with urlopen(img_url) as url:
        img = mpimg.imread(url, format='jpg')
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(caption)

plt.tight_layout()
plt.show()
```
