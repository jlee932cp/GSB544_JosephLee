1. Write a function that takes in information about a penguin, and returns one of the following definitions:

*   "Big Mouth Billy": Male penguins with bill length times bill depth greater than 800.
*   "Dainty Daisy": Female penguins with flipper length less than 5% of body mass.
*   "Average Adelie:" Any Adelie penguins that do not fall into either category.
*   "Missing": Penguins with necesary information missing to be categorized
*   "Other": Any penguins that do not fall into any of the categories.



```python
import pandas as pd
def assign_penguin_category(bill_length, bill_depth, flipper_length, body_mass, sex, species):
    """
    Categorize a penguin based on provided definitions.

    Parameters:
    penguin (dict): A dictionary containing penguin information.

    Returns:
    str: A string indicating the penguin's category.
    """
    # Check for missing values
    if pd.isna(bill_length) or pd.isna(bill_depth) or pd.isna(flipper_length) or pd.isna(body_mass) or pd.isna(sex) or pd.isna(species):
        return "Missing"

    # Debugging: Log the input values
    #print(f"Values: {bill_length}, {bill_depth}, {flipper_length}, {body_mass}, {sex}, {species}")

    # Big Mouth Billy
    if sex.strip().capitalize() == 'Male' and (bill_length * bill_depth) > 800:
        return "Big Mouth Billy"

    # Dainty Daisy
    if sex.strip().capitalize() == 'Female' and flipper_length < 0.05 * body_mass:
        return "Dainty Daisy"

    # Average Adelie
    if species.strip().capitalize() == 'Adelie' and not (sex.strip().capitalize() == 'Male' and (bill_length * bill_depth) > 800) and not (sex.strip().capitalize() == 'Female' and flipper_length < 0.05 * body_mass):
        return "Average Adelie"

    # Other
    return "Other"
```

2. Use an iterable function to create a new variable called `category_name` that adds these labels.

```python
penguins['category_name'] = penguins.apply(
    lambda row: assign_penguin_category(  # Ensure correct spelling here
        bill_length=row['bill_length_mm'],
        bill_depth=row['bill_depth_mm'],
        flipper_length=row['flipper_length_mm'],
        body_mass=row['body_mass_g'],
        sex=row['sex'],
        species=row['species']
    ),
    axis=1
)
```
